## 1. Variables and Data Types

```python
# Basic variable assignment
name = "Python"
age = 30
height = 5.9
is_active = True

print(f"Name: {name}, Type: {type(name)}")
print(f"Age: {age}, Type: {type(age)}")
print(f"Height: {height}, Type: {type(height)}")
print(f"Active: {is_active}, Type: {type(is_active)}")
```

## 2. Operators

```python
# Arithmetic operators
a = 10
b = 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Floor Division: {a} // {b} = {a // b}")
print(f"Modulo: {a} % {b} = {a % b}")
print(f"Exponent: {a} ** {b} = {a ** b}")
```

```python
# Comparison operators
x = 5
y = 8

print(f"{x} == {y}: {x == y}")
print(f"{x} != {y}: {x != y}")
print(f"{x} < {y}: {x < y}")
print(f"{x} > {y}: {x > y}")
print(f"{x} <= {y}: {x <= y}")
print(f"{x} >= {y}: {x >= y}")
```

## 3. Conditionals (if/elif/else)

```python
# Simple if-else
score = 75

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'F'

print(f"Score: {score}, Grade: {grade}")
```

## 4. Loops (for and while)

```python
# For loop
print("For loop (counting 1 to 5):")
for i in range(1, 6):
    print(f"Number: {i}")
```

```python
# While loop
print("\nWhile loop (countdown from 5):")
count = 5
while count > 0:
    print(f"Countdown: {count}")
    count -= 1
print("Blast off!")
```

1. Write a program that takes a number and checks if it's even or odd
2. Print the multiplication table for a given number
3. Calculate the sum of numbers from 1 to N using a loop

### Exercise 1 — Even or Odd (Try here)
Write a program that takes an integer and prints whether it's even or odd. Replace the sample code below with your attempt.

```python
# Student attempt: Even or Odd
try:
    num = int(input('Enter an integer: '))
    if num % 2 == 0:
        print(f'{num} is even')
    else:
        print(f'{num} is odd')
except ValueError:
    print('Please enter a valid integer')
```

### Exercise 2 — Multiplication Table (Try here)
Print the multiplication table (1..10) for a given integer. Replace the sample code below with your attempt.

```python
# Student attempt: multiplication table
try:
    n = int(input('Enter a number for its multiplication table: '))
    for i in range(1, 11):
        print(f'{n} x {i} = {n * i}')
except ValueError:
    print('Please enter a valid integer')
```

### Exercise 3 — Sum 1..N (Try here)
Calculate the sum of numbers from 1 to N using a loop. Replace the sample code below with your attempt.

```python
# Student attempt: sum from 1 to N
try:
    N = int(input('Enter N: '))
    total = 0
    for i in range(1, N+1):
        total += i
    print(f'Sum 1..{N} = {total}')
except ValueError:
    print('Please enter a valid integer')
```

### Solutions (reference)
Reference function-based solutions (use after attempting the exercises).

```python
def is_even(n):
    return n % 2 == 0

def multiplication_table(n, upto=10):
    return [n * i for i in range(1, upto+1)]

def sum_1_to_n(n):
    return n * (n + 1) // 2

# Quick demo (uncomment to test)
# print(is_even(4))
# print(multiplication_table(5))
# print(sum_1_to_n(10))
```
